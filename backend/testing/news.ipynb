```python
import requests
from pprint import pprint
from datetime import datetime
from sentiment import get_sentiment
import time
```

```python
url = "https://news.treeofalpha.com/api/news?limit=1000"

r = requests.get(url)
if r.status_code == 200:
    data = r.json()
```

```python
filteredData = []
for i in data:
    if i['source'] == 'Blogs' and str(i['title']).startswith("COINTELEGRAPH"):
        filteredData.append(i)
print(len(filteredData))
```

```python
filteredData
```

```python
for i in filteredData:
    # print(i['time'])
    publishTime = datetime.fromtimestamp(i['time']/1000)
    publishTime = publishTime.strftime("%H:%M %d-%m-%Y")
    title = i['title'][15:]
    sentiment = get_sentiment(title)
    # time.sleep(0.5)
    print(publishTime, title, sentiment)
```
